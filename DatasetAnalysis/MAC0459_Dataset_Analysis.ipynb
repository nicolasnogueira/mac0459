```python
%matplotlib inline 
import matplotlib.pyplot as plt

import warnings; warnings.simplefilter('ignore')
```

```python
import pandas as pd
import seaborn as sns
from statistics import mean, median

students_mat = pd.read_csv('./student/student-mat.csv')
students_por = pd.read_csv('./student/student-por.csv')
```

```python
students_mat.info()
```

```python
students_por.info()
```

## Tempo de viagem x Quantidade de faltas

```python
sns.distplot(students_mat['traveltime'], rug=True, kde=False)
sns.jointplot(students_mat['traveltime'], students_mat['absences'], kind='reg');
```

```python
sns.boxplot(x="traveltime", y="absences", data=students_mat)
```

```python
sns.distplot(students_por['traveltime'], rug=True, kde=False)
sns.jointplot(students_por['traveltime'], students_por['absences'], kind='reg');
```

```python
sns.boxplot(x="traveltime", y="absences", data=students_por)
```

## Em relacionamento romântico? x Horas de estudo por semana

```python
students_por['romantic'].value_counts()
```
